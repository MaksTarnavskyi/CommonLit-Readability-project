```python
import sys
import pandas as pd
from omegaconf import OmegaConf

sys.path.append('../')

from src.preparation.train_dev_split import split
from src.utils.plot_visualization import plot_histogram
```

```python
conf = OmegaConf.load('../configs/preparation/train_dev_split.yml')
OmegaConf.to_object(conf)
```

```python
data_folder_path = '../data/input/'
df = pd.read_csv(data_folder_path+'train.csv')
```

```python
train_df, dev_df = split(df, conf.test_size, conf.stratified, conf.target_column_name, conf.random_state)
```

```python
p = plot_histogram(df, conf.target_column_name, x_label='target value', y_label='frequency', title='Target value distribution')
```

```python
p = plot_histogram(
    train_df, conf.target_column_name, 
    x_label='target value', y_label='frequency', 
    title='Target value distribution in train part'
)
```
